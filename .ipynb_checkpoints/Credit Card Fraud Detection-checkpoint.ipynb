```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
predicted = model.predict(X_train)
print(predicted)
```

```python
acc = accuracy_score(Y_train, predicted)
print('accuracy of train data : ', acc)
```

```python
predicted_ = model.predict(X_test)
print(predicted_)
```

```python
acc_ = accuracy_score(Y_test, predicted_)
print('accuracy of test data : ', acc_)
```
